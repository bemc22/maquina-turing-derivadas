```python
def to_latex(input):
    if(input == 0):
        return "Ecuacion\:invalida"
    res = ''
    cont = 0
    j = ''
    for i in input:  
        if i == '1':
            cont += 1
        else:
            if j == '1':
                res += str(cont)
                res += i
                cont = 0
            else:
                res += i            
        j = i
        
    if(cont != 0):
        res += str(cont)
    return res
    
```

```python
from MTfunctions import *
from IPython.display import display, Markdown
from IPython.display import clear_output



def MTTM(myInput, tapes, numTapes):
    
    states = {'w0', 'q0', 'q1', 'q2', 'q3', 'q4','q5','q6','q7', 'q8', 'q9', 'q10', 'q11', 'q12', 'q13', 'q14', 'q15',
              'q16', 'c1', 'c2', 'd1', 'd2', 't1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9', 't10', 't11', 't12',
              't13', 't14', 't15'}
    input_symbols = {'x','^','{','}','1', '+', '-', 'M', 's', 'i', 'n', 'c', 'o', 't', 'a', 'n', '(', ')'}
    
    transitions = {
        'w0':{
            # Estado inicial 
            # Decidimos si vamos al estado de:
            # Estado empiza con menos
            # Monomio sin constante
            # Empieza con un coeficiente
            # Trigonometrica
            ('-',''):('q0', ('-','-'), ('r','r')),
            ('x',''):('q1', ('x','' ), ('r','s')),
            ('1',''):('q9', ('M','' ), ('r','s')),
            ('s',''):('t1', ('s','c'), ('r','r')),
            ('c',''):('t3', ('c','' ), ('s','l')),
            ('t',''):('t7', ('t','s'), ('r','r'))
        },
        
        'q0': {
            # Estado: empieza con menos
            # Decidimos si vamos al estado de:
            # Monomio sin constante
            # Monomio con constante
            # Trigonometrica
            ('x',''):('q1', ('x','' ), ('r','s')),
            ('1',''):('q9', ('M','' ), ('r','s')),
            ('s',''):('t1', ('s','c'), ('r','r')),
            ('c',''):('t3', ('c','' ), ('s','l')),
            ('t',''):('t7', ('t','s'), ('r','r'))
        },
        
        'q1': {
            # Estado: Monomio sin constante
            # Decidimos en funcion si es:
            # Monomio de grado (1)
            # Monomio de grado (2) o m√°s
            ('^',''):('q2', ('^','' ), ('r','s')),
            ('+',''):('q7', ('+','1'), ('s','r')),
            ('-',''):('q7', ('-','1'), ('s','r')),
            ('' ,''):('q7', ('' ,'1'), ('s','r'))
        },
        
        'q2': {
            # Se sigue la sintaxis de latex para un monomio
            ('{',''):('q3', ('{',''), ('r','s'))
        },
        
        'q3': {
            # Bucles dentro de exponente del monomio
            # Se escriben como coeficientes en la cinta (2)
            ('1',''):('q3', ('1','1'), ('r','r')),
            ('}',''):('q4', ('}','' ), ('l','s'))
        },       
   
        
         'q4': {
            # Regresamos al comienzo del monomio
            ('1',''):('q4', ('1',''), ('l','s')),
            ('{',''):('q4', ('{',''), ('l','s')),
            ('^',''):('q4', ('^',''), ('l','s')),
            ('x',''):('q4', ('x',''), ('l','s')),
            ('','' ):('q5', ('', ''), ('r','s')),
            ('+',''):('q5', ('+',''), ('r','s')),
            ('-',''):('q5', ('-',''), ('r','s'))
        },
        
        'q5': {
            # Escribimos el monomio en la cinta 2
            # Omitimos el primer 1
            ('x',''):('q5', ('x','x'), ('r','r')),
            ('^',''):('q5', ('^','^'), ('r','r')),
            ('{',''):('q5', ('{','{'), ('r','r')),
            ('1',''):('q6', ('1','' ), ('r','s'))
        },
        
        'q6': {
            # Leyendo el segundo (1) 
            # en el exponente de la cinta (1)
            # Escribimos el primer uno del exponente
            # de la cinta (2)
            ('1',''):('d1', ('1','1'), ('r','r')),
            ('}',''):('q7', ('}','}'), ('r','r'))
        },
        
        'd1': {
            # Si leemos otro 1 en la cinta (1)
            # Significa que es de grado (3) o m√°s
            # Si leemos un } en la cinta (1)
            # significa que es de grado (2)
            ('1',''):('q6',  ('1','1'), ('r','r')),
            ('}',''):('d2',  ('}', ''), ('s','l'))
     
        },
        
        'd2': {    
            # Al ser monomio de grado (2)
            # Procedemos a eliminar la sintaxis
            # (^{1}) de la cinta (2)
            # Con el fin de evitar redundancia
            ('}','1'):('d2', ('}','' ), ('s','l')),
            ('}','{'):('d2', ('}','' ), ('s','l')),
            ('}','^'):('d2', ('}','' ), ('s','l')),
            ('}','x'):('q7', ('}','x'), ('r','r')),
        },
        
        'q7': {
            # Este estado sirve como un estado de transicion
            # Para "reiniciar" nuestra maquina de turing
            # para cada termino de la expresion
            # En caso de no haber m√°s terminos 
            # vamos al estado final
            ('+',''):('q0', ('+','+'), ('r','r')),
            ('-',''):('q0', ('-','-'), ('r','r')),
            ('',''):('q8',  ('',''  ), ('s','s'))
        },
        
        'q8':{
            # Terminamos de leer la cinta (1)
            ('',''):('q8', ('',''), ('s','s'))
        },
        
        'q9': {
            # Estado: Empieza con un coeficiente
            # Decidimos si es:
            # Una constante
            # Monomio con constante
            ('1',''):('q9' , ('1',''), ('r','s')),
            ('x',''):('q10', ('x',''), ('r','s')),
            ('+',''):('c1' , ('+',''), ('l','s')),
            ('-',''):('c1' , ('-',''), ('l','s')),
            ('' ,''):('c1' , ('' ,''), ('l','s'))
            
        },
        
        'c1':{
            # Una constante
            # Borramos la marca
            # Escribimos 0 en la cinta (2)
            ('1',''):('c1' , ('1','') , ('l','s')),
            ('M',''):('c2', ('1','0') , ('r','r'))
        },
        
        
        'c2':{
            # Pasamos al siguiente termino
            # de la expresion
            ('1',''):('c2' , ('1','' ) , ('r','s')),
            ('+',''):('q0' , ('+','+') , ('r','r')),
            ('-',''):('q0' , ('-','-') , ('r','r')),
            ('','' ):('q8' , ('' ,'' ) , ('s','s'))
        },
        
        
        'q10':{
            # Monomio con constante
            # Decidimos si es:
            # Monomio de grado (1)
            # Monomio de grado (2) o m√°s
            ('^',''):('q11', ('^',''), ('r','s')),
            ('+',''):('q15', ('+',''), ('l','s')),
            ('-',''):('q15', ('-',''), ('l','s')),
            ('','' ):('q15', ('' ,''), ('l','s'))
        },
        
        'q11':{
            # Monomio de grado (2) o m√°s
            # Entramos en el exponente
            ('{',''):('q12', ('{',''), ('r','s'))
        },
        
        'q12':{
            # Escribimos el exponente
            # Como un coeficiente en la cinta (2)
            ('1',''):('q12', ('1','1'), ('r','r')),
            ('}',''):('q13', ('}',''), ('l','s'))
        },
        
        'q13':{
            # Regresamos al coeficiente 
            # del monomio reemplazamos la marca
            # por un (1)
            ('1',''):('q13', ('1',''), ('l','s')),
            ('{',''):('q13', ('{',''), ('l','s')),
            ('^',''):('q13', ('^',''), ('l','s')),
            ('x',''):('q13', ('x',''), ('l','s')),
            ('M',''):('q14', ('1',''), ('r','s'))
        },
        
        'q14':{
            # Pueden pasar dos cosas:
            #
            #  1. Se encuentra una x luego
            #  significa que terminamos 
            #  con la multiplicacion
            #  
            #  2. Encontramos otro (1) en la
            #  cinta (1) luego continuamos
            #  con la multiplicaci√≥n
            #
            ('x',''):('q5', ('x','x'), ('r','r')),
            ('1',''):('q9', ('M',''), ('r','s'))
        },
        
        'q15':{
            # Monomio de grado (1)
            # Reemplazamos la marca por un (1)
            ('x',''):('q15', ('x','' ), ('l','r')),
            ('1',''):('q15', ('1','' ), ('l','s')),
            ('M',''):('q16', ('1','1'), ('r','r'))
        },
        
        'q16':{
            # Recorremos todo el termino
            # y pasamos al estado de transicion q7
            ('1',''):('q16', ('1','1'), ('r','r')),
            ('x',''):('q7' , ('x','' ), ('r','s'))
        },
        
        # Estados para las funciones trigonometricas 
        # en este caso la MT funciona como un traductor
        #
        # Si leemos sin en la cinta (1)
        # escribimos cos en la cinta (2)
        #
        #
        # Si leemos cos en la cinta (1)
        # escribimos - sin en la cinta (2)
        #
        #
        # Si leemos tan en la cinta (1)
        # escribimos sec en la cinta (2)
        #
        
        't1':{
            # Empezamos escribir la derivada del sin(x)
            ('i',''):('t2', ('i','o'), ('r','r'))
        },
        
        't2':{
            ('n',''):('t13', ('n','s'), ('r','r'))
        },
        
        't3':{
            # Empezamos a escribir la derivada para el cos(x)
            # En funcion del signo anterior al cos(x)
            # Cambiamos el signo por su opuesto
            ('c','+'):('t4', ('c','-'), ('s','r')),
            ('c','-'):('t4', ('c','+'), ('s','r')),
            ('c','') :('t4', ('c','-'), ('s','r'))
        },
        
        't4':{
            # Empezamos a reescribir el cos(x) como sin(x)
            ('c',''):('t5', ('c','s'), ('r','r'))
        },
        
        't5':{
            ('o',''):('t6', ('o','i'), ('r','r'))
        },
        
        't6':{
            # Terminamos de escribir la derivada del cos(x)
            ('s',''):('t13', ('s','n'), ('r','r'))
        },
        
        't7':{
             # Empezamos a escribir la deriva de tan(x)
            ('a',''):('t8', ('a','e'), ('r','r'))
        },
        
        't8':{
            ('n',''):('t9', ('n','c'), ('r','r'))
        },
        
        't9':{
            ('(',''):('t10', ('(','^'), ('s','r'))
        },
        
        't10':{
            ('(',''):('t11', ('(','{'), ('s','r'))
        },
        
        't11':{
            ('(',''):('t12', ('(','2'), ('s','r'))
        },
        
        't12':{
            # Terminamos de escribir la derivada de tan(x)
            ('(',''):('t13', ('(','}'), ('s','r'))
        },
        
        # Las 3 funciones sin(x) cos(x) tan(x)
        # terminan ambas con "(x)"
        # luego estos estamos se comparten entre ellas
        # t13, t14 y t5
        
        't13':{
            ('(',''):('t14', ('(','('), ('r','r'))
        },
        
        't14':{
            ('x',''):('t15', ('x','x'), ('r','r'))
        },
        
        't15':{
            # Al terminar de derivar la funcion
            # trigonometrica respectiva
            # prodedemos a pasar
            #  al estado de transici√≥n q7
            (')',''):('q7', (')',')'), ('r','r'))
        }
    }
    
    initial_state = 'w0'
    final_states = {'q8'}
    
    cinta2 = validate(tapes,numTapes,myInput,input_symbols,states,transitions,initial_state,final_states)
    ecuacion = to_latex(myInput)
    derivada = to_latex(cinta2)
    display(Markdown(rf"""
$f(x) = {ecuacion}$

$f'(x) = {derivada}$
    """))

    
    #plt.axis('off')
```
